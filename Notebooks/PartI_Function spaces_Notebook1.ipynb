If we are to develop algorithms for the problems of continuum mathematics, for finding solutions to differential equations, determining curves or surfaces optimizing an objective functional, or determining an equilibrium probability distribution, then we must learn about functions and how they can be represented in a computer. As computers are discrete entities and functions are typically continuous objects, this means our representations will have be approximations. Accordingly, we must learn how functions can be approximated and thus we are going to study the geometry of functional spaces.

## Definitions:

The classes of functions we will consider will be vector spaces or subsets of a vector space. These spaces have a notion of lenght for their vectors.

**Norms**:
A norm in a vector space $V$ is a function $n:V\to\mathbb{R}$ with the following properties

1.(Positivity) $n(x)\geq 0$ for all $x\in V$ and $n(x)=0$ if and only if $x=0$.

2.(Homogeneity) $n(\lambda x) = |\lambda| n(x)$ for all $x\in V$ and $\lambda\in\mathbb{R}$.

3.(Triangle inequality) $n(x+y)\leq n(x)+n(y)$ for all $x,y\in V$.

These properties guarantee that the function $d(x,y) := n(x-y)$ defines a distance in $V$ turning it into a **metric space**.

**Banach space**:
This is a normed space $(V,n)$ where the metric $d(x,y) = n(x-y)$ is complete. 

**Hilbert space**:
This is a vector space $H$ with an inner product $(x,y)$ with the property that the corresponding norm

$$ \|x\| := \sqrt{(x,x)},\; d(x,y) := \|x-y\|$$

make $H$ into a separable Banach space. 


## Basic examples

Here is a list of important normed spaces: 

* The real and complex vector spaces $\mathbb{R}^n$, $\mathbb{C}^n$
* The space of real $m\times n$ matrices
* The space of real valued continuous functions in a compact set $\Omega \subset \mathbb{R}^d$, denoted $C^0(\Omega)$.
* The space of integrable functions on the real line, $L^1(\mathbb{R})$.
* The space of square-integrable functions in $(0,1)$, $L^2(0,1)$.
* The space of square-summable fiunctions over $\mathbb{Z}$, denoted $\ell^2(\mathbb{Z})$. This can be thought of as the space of all doubly infinite sequences $x= (a_n)_{n\in\mathbb{Z}}$ such that $$ \sum \limits_{n} |a_n|^2 < \infty $$
* The space of real-valued functions on a graph $G$, denoted $C(G)$ or $\mathbb{R}^G$. A particular example is the space of functions over the vertices of the $n$-dimensional unit cube, $\{0,1\}^n$.
* Other spaces we will see later are Sobolev space and spaces of measures. 

### Polynomials

The class of all polynomials of degree $\leq N$ form a subspace of $C([a,b])$, it has dimension $N+1$. A basis for this subspace is given by the monomials

$$  x^0,x^1,\ldots,x^N $$

A crucial theorem is Weierstrass theorem.

**Theorem**: Given a continuous function $f$ in $[a,b]$ and $\varepsilon>0$ there is some polynomial $p$ such that $ \max \limits_{a\leq x \leq b} |f(x)-p(x)| < \varepsilon$.

We will discuss a proof in class. 


**Bernstein basis.** Weierstrass original proof was not constructive. A constructive proof was done by Bernstein, who introduced a class of polynomials that is now named after him:

Given $n$, the Bernstein of polynomials of degree $n$ are the $n+1$ polynomials $B_{k,n}$ ($k=0,\ldots,n$) given by

$$B_{k,n}(x) =  {n \choose k}   x^k(1-x)^{n-k}  $$

Then, given $f \in C([0,1])$ we define a polynomial for each $n$,

$$B_n(f)(x) = \sum \limits_{k=0}^n f(k/n) B_{k,n}(x) $$

Bernstein showed, using probabilistic ideas, that

$$\max \limits_{0\leq x \leq 1}|f(x)-B_n(f)(x)| \to 0 \text{ as } n \to \infty $$

which provided a constructive improvement on Weierstrass theorem.

**Legendre basis.** A different basis for the space of polynomials of degree $n$ was introduced by Legendre, based on the idea of orthogonality. One considers the inner product

$$(f,g) = \int_0^1 f(x)g(x)\;dx $$

and one applies the Gram-Schmidt process to the family of polynomaisl $x^0,x^1,x^2,\ldots$, and normalizes the resulting polynomials so they all equal $1$ at $x=1$ one arises at the Legendre polynomials$P_n$ $n=0,1,\ldots$. Each polynomial $P_n$ is of degree $n$ and for every $n$ and $m$ with $n\neq m$ one has

$$ \int_0^1 P_n(x)P_m(x)\;dx = 0$$


### Trigonometric Polynomials

A trigonometric polynomial of period $\tau$ is a function that can be expressed as

$$ f(x) = a_0 + \sum \limits_{k=1}^N a_k \sin(\tfrac{2\pi k}{\tau} x)+b_k \cos(\tfrac{2\pi k}{\tau} x)$$

In particular, such a function is a sum of finitely many sines and cosines with a common period (that is, $\tau$). The number $N$ is called the degree of the polynomial. An important fact (that we will discuss later on) is that trigonometric functions form an orthonormal basis for the Hilbert space $L^2(0,1)$. Also important is that every function in $C([0,1])$ can be approximated uniformly by a trigonometric polynomial -- that is, trigonometric polynomials are dense in $C([0,1])$.

### Piecewise linear functions

Given the points $0=x_0<x_1<\ldots<x_N=1$ one can consider the subspace of $C([0,1])$ made out of functions which are affine in each of the intervals $(x_k,x_{k+1})$ -- in particular, the graph of such functions are polygonal curves. In this context the set of points $\{x_0,x_1,\ldots,x_N\}$ determines the values of such a function entirely. The resulting class of functions is called Piecewise linear functions. We will see later that also any function in $C([0,1])$ can be approximated by piecewise linear functions.

Piecewise linear functions are also used in higher dimensions, there what plays the role of the points $\{x_0,x_1,\ldots,x_N\}$ is a triangulation. Just as in 1D, piecewise linear functions can also approximate continuous functions in higher dimensions. 

### Single layer neural networks

The logit function is the function

$$ \sigma(t) = \frac{1}{1+e^{-t}}$$

This function takes values in $(0,1)$, converging to $0$ as $t \to -\infty$ and to $1$ as $t \to + \infty$. If we fix $N$, then given vectors $a_1,\ldots,a_N \in \mathbb{R}^d$, $b_1,\ldots,b_N\in\mathbb{R}$, and numbers $c_1,\ldots,c_N$ we can form a real valued function in $\mathbb{R}^d$, 

$$ f(x) = \sum \limits_{k=1}^N c_k \sigma(a_k\cdot x+b_k)$$

Each selection of the paramaters $a_k,b_k,c_k$ (also called weights) gives rise to a continuous function. This structure is known as a feedforward neural network. A Theorem of Cybenko says that given any continuous function in a bounded domain of $\mathbb{R}^d$ one can approximate uniformly using feedforward neural networks, with the number of terms $N$ increasing the smaller we want the error to be.


## Convolutions

The convolution of two functions $f,g:\mathbb{R}\to\mathbb{R}$ is a new function $f*g$ given by

$$(f*g)(x) := \int_{-\infty}^{\infty}f(y)g(x-y)\;dy $$

In general, if one convolves $f$ with some function $g$, the resulting function will have better smoothness than either $f$ or $g$. This is because the convolution $f*g$ is returning averaged values of $f$ weighted against $g$. For example, if 

$$g(x) = \left \{ \begin{array}{ll} \frac{1}{2L} \text{ if } |x|<L \\ 0 \text{ otherwise } \end{array}\right.$$

Then for every $x$, $f*g(x)$ is equal to the average of $f$ over the interval $(x-L,x+L)$.

For periodic functions, there is a different notion of convolution. If $f,g$ are $1$-periodic functions, then their convolution is defined as

$$(f*g)(x) = \int_0^1 f(y)g(x-y)\;dy $$

# Fourier Series

Definition:
Given an **integrable function**\* $f:[0,1]\to \mathbb{C}$ its *Fourier Series* is the representation of $f$ in the form

$$ f(x) = \sum \limits_{n \in \mathbb{Z}} c_n e^{2\pi i nx},\;\;c_n := \int_0^1 f(y)e^{-2\pi iny}\;dy$$

The limit of the series being understood in some weak sense we will describe later. When this holds the sequence of coefficients $\{c_n\}_n$ are called the Fourier Coefficients of the function.

\*if you prefer, for simplicity, you may assume this function is continuous save for an at most finite number of jump discontinuities 

The partial sums of $f$ are the function $S_N(f)$ defined by

$$ S_N(f)(x) = \sum \limits_{n = - N}^{n=N} c_n e^{2\pi i nx} $$

The coefficients $c_n$ being the same as before. The equality between $f$ and its Fourier Series must be understood in the sense that, if $f \in L^2(0,1)$, then

$$ \lim \limits_{N \to \infty}\|S_N(f)-f\|_{L^2} = 0$$

We will see in the exercises some general classes of functions $f$ such that the Fourier Series converges pointwise for every $x$ -- and even uniformly in $x$.
