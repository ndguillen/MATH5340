    1. There is a matrix $A=A_h$ and a matrix $B=B_h$ such that for every $n$ we have $\hat x_{n+1} = A x_n$ and $\check x_{n+1} = Bx_n$. Write down each of these matrices explicitly in terms of $h$.

## Part 3. Graphs, Laplacians, and Markov chains.

1. Consider a graph with vertices $G=\{x_1,\ldots,x_N\}$ and symmetric weight matrix $w_{ij}$. A function $f:G\to\mathbb{R}$ is said to have a *strict* local maximum at a point $x_{i_0} \in G$ if $f(x_{i_0})>f(x_j)$ for every $j$ such that $x_i$ and $x_j$ are neighbors. Show that if $f$ has a strict local maximum at $x_{i_0}$ then the graph Laplacian of $f$ at $x_{i_0}$ needs to be strictly negative.
2. Fix $h>0$ and consider the graph $G = \{x_1,\ldots,x_N\}$ given by the intersection of $h\mathbb{Z}^2$ with the unit square $[0,1]^2$, and where the weights $w_{ij}$ are defined to be $0$ if $x_i$, $x_j$ are not neighbors in $h\mathbb{Z}^2$ and $1/(4h^2)$ if they are neighbors. Compute the graph Laplacian of the function $f(x_i) := x_i \cdot e$ where $e$ is some given vector in $\mathbb{R}^2$.


## Part 4. Finite elements.


1. Show that if $A$ is a $N\times N$ matrix, then $x$ solves the equation $$Ax =b$$ *if and only if* $$Ax \cdot y = b\cdot y \text{ for all } y \in \mathbb{R}^N$$ What does this have to do with the finite element method?
2. Consider $f(x,y)$, a twice differentiable function of two variables, and $\bar v(x,y)$ the vector field $\bar v(x,y) = (-y,x)$. Show that $f$ solves $$ \Delta f + \nabla f \cdot \bar v = 0$$ in the disc $D$ if and only if $$-\int_D \nabla f\cdot \nabla \phi + f \nabla \phi \cdot \bar v \;dxdy = 0 \text{ for all } \phi \in C^2_c(D)$$
3. The following illustrates FEM in the 1D case. Fix $N$. Denote by $\mathcal{S}$ the subset of $C([0,1])$ of functions which are affine when restricted to every interval of the form $[k/N,(k+1)/N]$.
