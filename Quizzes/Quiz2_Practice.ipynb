# Quiz 2 practice problems

1. Consider the forward Euler scheme for the differential equation $$\dot x(t) = Ax(t)$$ where $x(t) \in \mathbb{R}^d$ and $A$ is a $d\times d$ matrix, and the time step is $h>0$. Let $x_n$ be a sequence generated by the scheme, starting from some $x_0$. Prove that for every $n$ we have $$x_n = (I+hA)^nx_0$$ Then, assuming that $A$ is positive definite, show that for every fixed $h>0$ we have $\|x_n\|_2 \to \infty$ as $n\to\infty$.  
  Hint: Since $A$ is positive definite there is a basis of eigenvectors $e_1,\ldots,e_d$ with $Ae_k = \lambda_k e_k$ and $0<\lambda_1\leq \lambda_2 \leq \ldots \leq \lambda_d$, what happens if $x_0 = e_k$ for some $k$? What happens in general, expressing $x_0$ in this basis and multiplying by $(I+hA)$ $n$ consecutive times?
2. Suppose that $f:\mathbb{R}^d\to\mathbb{R}^d$ is a Lipschitz function with Lipschitz constant $L>0$ and $x_* \in\mathbb{R}^d$. Show for what values of $h$ we can guarantee the map $T:\mathbb{R}^d\to\mathbb{R}^d$ given by $T(x) = x_* + hf(x)$ is a contraction mapping, and explain the relevance of this observation to the backward Euler scheme. 
3. The **trapezoidal rule** refers to a numerical scheme for $\dot x(t) = f(x(t))$ where, for time step $h>0$ and initial condition $x_0$ the sequence $x_n$ is generated according to the implicit iterative relation $$x_{n+1}-x_n = \frac{h}{2}(f(x_n)+f(x_{n+1}))$$ Show for what values of $h>$ can we guarantee that the map given by $T(x) = x_* + \frac{h}{2}(f(x_*) +f(x))$ is a contraction mapping. Then, show that when $f(x) = Ax$ with $A$ a $d\times d$ matrix, we have the relation $$ x_{n} = \left [(I-\frac{h}{2}A)^{-1}(I+\frac{h}{2}A)\right ]^nx_0  $$ provided $h$ is sufficiently small. Hint: You may use that $I-\frac{h}{2}A$ is invertible if $h\|A\|<2$, where $\|A\| = \max_{x \neq 0} \{\|Ax\|/\|x\|\}$.
4. The local truncation error for the trapezoidal rule applied to $\dot x(t) = f(x(t))$ is defined by $$\varepsilon_n^{(h)} = \frac{1}{h}\left [x(t_{n+1})-x(t_n) - \frac{h}{2}(f(x(t_{n+1}))+f(x(t_n))) \right ] $$ where $x(t)$ is the analytical solution to the differential equation and $t_n = nh$ ($h>0$ is the time step). Consider the 1d equation $\dot x = -\lambda x$, $\lambda>0$, with initial data $x_0$ and for times $t>0$. Show that in this case $$|\varepsilon_n^{(h)}| \leq h^{-1}|x_0|\left [ e^{-h\lambda}-1 +\frac{\lambda h}{2}(e^{-h\lambda}+1 )\right ] $$ 
5. Show that the map $T:C([0,\tau])\to C([0,\tau])$ defined by $$(Tx)(t) = \int_0^t e^{-x(s)}\;ds$$ is a contraction mapping for every $\tau \in (0,1)$.
